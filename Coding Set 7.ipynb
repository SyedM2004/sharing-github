```python
def main():
  #Define how many days are needed, in this case its 7 days
  NUM_DAYS = 7
  #make a list for the days
  sales = [0] * NUM_DAYS

  print("enter the sales for each day")

  #create an input for the sales made in each day
  for index in range(len(sales)):
    sales[index] = float(input(f'day #{index + 1}: '))

  #make sure the numbers inputted are seen
  print('Here are the values you entered: ')
  for value in sales:
    print(value)

    #create a code that shows the sum
    total_sales = sum(sales)
    print(f'total sales: {total_sales}')

  #call the function into play
if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt

def main():
    # First we list expenses
    expenses = {
        "rent": 900,
        "Gas": 200,
        "food": 140,
        'clothing': 100,
        'car payment': 125,
        'misc': 100
    }

    file_path = "c:/path/to/your/directory/expenses.txt"

    # open the file being used
    with open('expenses.txt', 'w') as file:
        for category, amount in expenses.items():
            file.write(f"{category}: {amount}\n")

    categories = []
    amounts = []
    with open('expenses.txt', 'r') as file:
        for line in file:
            category, amount = line.strip().split(':')
            categories.append(category)
            amounts.append(float(amount))
    
    # we are going to use a pie chart for this so we have to create it.
   
    plt.figure(figsize=(8, 8))
    plt.pie(amounts, labels=categories, autopct='%1.1f%%', startangle=140)
    plt.title('monthly expenses')
    plt.axis('equal')

    plt.savefig('expense_pie_chart.png')

    plt.show()

    #Call the function into play
    #really hope this works on the first try
if __name__ == "__main__":
    main()
```

## Question 4

```python
#make an empty list to start off
numbers = []

#use the range function to get some numbers
for i in range(1, 21):
    while True:
        try:
            number = float(input(f"Enter number {i}: "))
            numbers.append(number)
            break
        except ValueError:
            print("Invalid input. Please enter a valid number.")

#calculate 
lowest = min(numbers)
highest = max(numbers)
total = sum(numbers)
average = total / len(numbers)

# print the numbers on the list from lowest to highest, with the average and total

print(f"Lowest number in the list: {lowest}")
print(f"Highest number in the list: {highest}")
print(f"Total of the numbers in the list: {total}")
print(f"Average of the numbers in the list: {average}")
```
